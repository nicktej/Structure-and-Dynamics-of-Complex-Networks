```python
word_assoc = networkx.read_edgelist('word_association_graph_DSF.txt', create_using=networkx.DiGraph(), nodetype=str, data=[('weight', float)] )
```

```python
weights = [d['weight'] for a,b,d in word_assoc.edges(data=True) ]

_ = plt.hist(weights, bins=100, log=True)
```

```python
#average_out

av_out = (sum(weights)/len(weights))
ave_out = []
out_deg_nodes = list(word_assoc.out_degree().values())

for x in range(len(out_deg_nodes)):
    ave_out.append(out_deg_nodes[x]*av_out)

_ = plt.hist(ave_out, bins = 32, log=True, color='green')
```

```python
#average_in

av_in = (sum(weights)/len(weights))
ave_in = []
in_deg_nodes = list(word_assoc.in_degree().values())

for x in range(len(in_deg_nodes)):
    ave_in.append(in_deg_nodes[x]*av_in)

plt.xscale('log')
_ = plt.hist(ave_in, bins = 320, log=True, color='red')
```

```python
#random_in

import random
random.shuffle(weights)

in_deg_weights = []
in_deg_nodes = list(word_assoc.in_degree().values())
counter = 0

for x in range(len(in_deg_nodes)):
    weightsum = 0
    for y in range(in_deg_nodes[x]):
        weightsum = weightsum + weights[counter]
        counter = counter + 1
    in_deg_weights.append(weightsum)
    
plt.xscale('log')
_ = plt.hist(in_deg_weights, bins = 320, log=True, color='purple')
```

```python
#random_out

import random
random.shuffle(weights)

out_deg_weights = []
out_deg_nodes = list(word_assoc.out_degree().values())
counter = 0

for x in range(len(out_deg_nodes)):
    weightsum = 0
    for y in range(out_deg_nodes[x]):
        weightsum = weightsum + weights[counter]
        counter = counter + 1
    out_deg_weights.append(weightsum)

_ = plt.hist(out_deg_weights, bins = 32, log=True)
```

```python
# The differences between the average_out graph and the random_out graph is slope of the graph and the x-axis values
# The x-axis in the random_out graph had a larger range in values (0-3.0), whereas for the average_out graph it is (0=2.0)
# The random_out graph is more right-skewed where as the average_out graph is more 'bell-shaped', as in more even compared to the random_out graph
# The random_out graph has a peak of roughly 10^2.5 at around 0.5-0.7 strength, whereas the average_out graph has a peak of approx. 10^2.5 at 1 strength

# The differences between the average_in graph and the random_in graph is towards the end of the x-axis
# The end of the random_in graph has smaller values compared to the average_in graph, but besides that they are pretty similar
```
